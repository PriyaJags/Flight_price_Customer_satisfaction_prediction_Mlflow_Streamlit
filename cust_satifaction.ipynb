Demographic Information:

    On-board service: Quality of on-board service.
    Leg room service: Leg room available.
    Baggage handling: Quality of baggage handling.
    Checkin service: Efficiency of check-in service.
    Inflight service: Quality of inflight service.
    Cleanliness: Cleanliness of the aircraft.

Flight Information:

    Flight Distance: Distance of the flight.
    Departure Delay in Minutes: Departure delay in minutes.
    Arrival Delay in Minutes: Arrival delay in minutes.
    
Target Variable:

satisfaction: Overall satisfaction of the passenger (satisfied/neutral or dissatisfied).

Age Distribution: Most passengers are aged between 27 and 51, with an average age of 39. This suggests a focus on a potentially working/middle-aged demographic.

Flight Distance: Flight distances vary widely, with the majority below 1743 miles. This indicates a mix of short-haul and potentially some medium-haul flights in the dataset.

Service Ratings: Ratings for inflight services, cleanliness, and comfort are distributed across the 0-5 scale, with medians generally above 3. This suggests generally positive ratings, but there's room for improvement. Further investigation into the distribution of lower ratings is warranted.

Delays: There are some long delays, but most passengers experience minimal or no delay. This suggests that while delays are a factor, they are not the norm. Outliers (long delays) might need special handling.

Satisfaction Rate: The satisfaction rate is roughly 43%, indicating a class imbalance (57% of passengers are dissatisfied or neutral). This class imbalance will need to be addressed during model training to avoid biased predictions.

```python
#Data cleaning
#Dropping unwanted columns like "Unnamed" ,"id"
df.drop(["Unnamed: 0", 'id'], axis=1, inplace=True)
```

```python
# Fill missing values in 'Arrival Delay in Minutes' with the median
df['Arrival Delay in Minutes'].fillna(df['Arrival Delay in Minutes'].median(), inplace=True)
```

```python
df.isna().sum()
```

```python
# encoding satisfaction because it's a binary classification target.--1 for satisfied and 0 for nuetral or dissatisfied
df['satisfaction'] = df['satisfaction'].apply(lambda x: 1 if x == 'satisfied' else 0)
```

```python
#Univariate Distribution Analysis with Respect to Satisfaction
import matplotlib.pyplot as plt
import seaborn as sns
# Define palette
palette = {0: "red", 1: "blue"}

# Get numeric columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Remove 'satisfaction' if present in numeric list
#if 'satisfaction' in numerical_cols:
 #   numerical_cols.remove('satisfaction')

# Plot
plt.figure(figsize=(24, 20))

for i, col in enumerate(numerical_cols):
    plt.subplot(6, 4, i+1)
    sns.kdeplot(df[col][df['satisfaction'] == 1], fill=True, label='Satisfied', color=palette[1])
    sns.kdeplot(df[col][df['satisfaction'] == 0], fill=True, label='Neutral or Dissatisfied', color=palette[0])
    plt.title(col)
    plt.xlabel('')
    plt.legend()

plt.tight_layout()
plt.show()
```

univariate analysis with comparison across a binary target variable, specifically using Kernel Density Estimation (KDE) plots to visualize the distribution of individual numeric features.
 Each subplot shows the distribution of a specific numerical feature, segmented by the target variable satisfaction (1 = Satisfied, 0 = Neutral or Dissatisfied).

The blue curve represents the distribution of values for satisfied passengers.

The red curve represents the distribution for neutral or dissatisfied passengers.

These plots help identify how each feature is distributed and whether the distributions differ significantly between the two satisfaction groups.

```python
#correlation heatmap for all features
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
corr = df[numerical_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

Positive Correlations with Satisfaction: Features like Online boarding, Inflight entertainment, and Seat comfort show a moderate positive correlation with passenger satisfaction. This indicates that passengers who rate these aspects of their flight experience highly are more likely to be satisfied overall. This is valuable information for airlines looking to improve customer satisfaction.

```python
#checking for chi-square test to check gender with satifaction. 
from scipy.stats import chi2_contingency

# Create contingency table
gender_table = pd.crosstab(df['Gender'], df['satisfaction'])

# Perform chi-square test
chi2, p, dof, expected = chi2_contingency(gender_table)
print(f"Chi-square statistic = {chi2}, p-value = {p}")
```
